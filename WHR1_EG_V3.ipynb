```python
# Begin SQL Alchemy Connection

engine = create_engine("sqlite:///world_happiness.sqlite")

# reflect an existing database into a new model
Base = automap_base()
# reflect the tables
Base.prepare(engine, reflect=True)

# Save reference for the table 
World_Happiness = Base.classes
```

```python
# Create our session (link) from Python to the DB
session = Session(engine)
```

```python
# Convert the database to our final dataframe for machine learning
Final_df = pd.read_sql_table('worldhappiness', engine)

# Set Country name as index
Final_df=Final_df.set_index('Country name')
Final_df.head()
```
